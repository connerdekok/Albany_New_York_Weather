```python
# reading csv into pandas dataframe
ny_data_csv = pd.read_csv("data/daily_data.csv")

ny_weather_df = pd.DataFrame(ny_data_csv)

# selecting columns used for analysis
ny_weather_clean_df  = ny_weather_df[["DATE", "DailyPeakWindDirection", "DailyPeakWindSpeed", "DailyPrecipitation", "DailySnowDepth", "DailySnowfall", "DailySustainedWindDirection", "DailySustainedWindSpeed", "DailyAverageDryBulbTemperature", "DailyMaximumDryBulbTemperature","DailyMinimumDryBulbTemperature"]]

ny_weather_clean_df
```

```python
ny_weather_clean_df = ny_weather_clean_df.astype({"DATE": "datetime64[ns]"})

ny_weather_clean_df.info()
```

```python
ny_weather_clean_sorted_df = ny_weather_clean_df.sort_values(by=['DATE'])
ny_weather_clean_sorted_df.head(500)
```

```python
ny_weather_clean_sorted_df.info()
```

```python
# Create a new column for 'Year' and 'Month' by extracting it from the 'Date' column

ny_weather_clean_sorted_df['Year'] = ny_weather_clean_sorted_df['DATE'].dt.year

ny_weather_clean_sorted_df['Month'] = ny_weather_clean_sorted_df['DATE'].dt.month

ny_weather_clean_sorted_df['Day'] = ny_weather_clean_sorted_df['DATE'].dt.day


ny_weather_clean_sorted_df
```

```python
# replacing the T (trace value) with 0
ny_weather_clean_sorted_df["DailyPrecipitation"] = ny_weather_clean_sorted_df["DailyPrecipitation"].replace("T", "0.0")
ny_weather_clean_sorted_df["DailySnowfall"] = ny_weather_clean_sorted_df["DailySnowfall"].replace("T", "0.0")
ny_weather_clean_sorted_df["DailySnowDepth"] = ny_weather_clean_sorted_df["DailySnowDepth"].replace("T", "0.0")

ny_weather_clean_sorted_df
```

```python
ny_weather_clean_sorted_df.info()
```

```python
ny_weather_clean_sorted_df = ny_weather_clean_sorted_df.astype({"DailyPrecipitation": "float", "DailySnowDepth": "float", "DailySnowfall": "float"})

ny_weather_clean_sorted_df.to_csv('ny_weather_clean.csv', index=False)

```

```python
conn = sqlite3.connect('data/daily_data.db')
ny_weather_clean_sorted_df.to_sql('new_york_weather', conn, if_exists='replace', index=False)

query = ('SELECT * FROM new_york_weather')

ny_weather_clean_sorted_df = pd.read_sql(query, conn)

ny_weather_clean_sorted_df
```

```python
# Create a 'Season' column based on the month
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'
```

```python
# Function to create the 'Season' column
ny_weather_clean_df['Season'] = ny_weather_clean_df['DATE'].dt.month.apply(get_season)

print(ny_weather_clean_df['Season'].head())  # Check if 'Season' column is there
numeric_columns = ['DailyMaximumDryBulbTemperature', 'DailyMinimumDryBulbTemperature', 
                   'DailySnowfall', 'DailyPeakWindSpeed']

# Group data by season and calculate the mean for the selected numeric columns
seasonal_data = ny_weather_clean_df.groupby('Season')[numeric_columns].mean()
```

```python
# Average Max and Min Temperatures Across Seasons 
# This chart shows how maximum and minimum temperatures vary across Winter, Spring, Summer, and Fall.
# Plot line charts comparing weather variables across seasons
plt.figure(figsize=(12, 8))

# Line chart for average temperature by season
plt.subplot(2, 1, 1)
plt.plot(seasonal_data.index, seasonal_data['DailyMaximumDryBulbTemperature'], label='Max Temperature', marker='o', color='red')
plt.plot(seasonal_data.index, seasonal_data['DailyMinimumDryBulbTemperature'], label='Min Temperature', marker='o', color='blue')
plt.title('Average Max and Min Temperatures Across Seasons')
plt.xlabel('Season')
plt.ylabel('Temperature (°F)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Average Snowfall and Peak Wind Speed Across Seasons.
# This chart shows the average snowfall and peak wind speed for each season.
# Line chart for snowfall and wind speed by season
plt.subplot(2, 1, 2)
plt.plot(seasonal_data.index, seasonal_data['DailySnowfall'], label='Snowfall (inches)', marker='o', color='purple')
plt.plot(seasonal_data.index, seasonal_data['DailyPeakWindSpeed'], label='Peak Wind Speed (mph)', marker='o', color='green')
plt.title('Average Snowfall and Peak Wind Speed Across Seasons')
plt.xlabel('Season')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
